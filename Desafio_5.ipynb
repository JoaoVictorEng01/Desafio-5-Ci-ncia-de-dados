```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Planilha do desafio 5 - Ciência de dados - Relatório do Banco x.csv')
data.columns = data.columns.str.upper()
data.head()
```

*2. Corrigir os dados faltantes e preencher as lacunas com valores adequados*  ✅

```python
data.isnull().sum()
# média para colunas numéricas e moda para colunas categóricas
for column in data.columns:
    if data[column].dtype == 'object':
        data[column].fillna(data[column].mode()[0], inplace=True)
    else:
        data[column].fillna(data[column].mean(), inplace=True)
data.head(100)

```

*3. Eliminar os outliers, atribuindo no lugar deles os valores adequados*

```python
threshold = 3
numeric_cols = data.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    col_zscore = (data[col] - data[col].mean()) / data[col].std()
    outliers = np.abs(col_zscore) > threshold
    mean_value = data[col].mean()
    if pd.api.types.is_integer_dtype(data[col]):
        mean_value = int(mean_value)
    data.loc[outliers, col] = mean_value

data.head(100)
```

*4. Eliminar os duplicados, para garantir a unicidade das informações*

```python
data.drop_duplicates(inplace=True)
```
